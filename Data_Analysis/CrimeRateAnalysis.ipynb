```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data['States/UTs'].nunique()
```

```python
data['States/UTs'].value_counts()
```

```python
count=data['States/UTs'].value_counts()
plt.figure(figsize=(26,16))
sns.barplot(count.index,count.values, alpha=0.8)
plt.title('States/UTs & Number of Crime reported', fontsize=20)
plt.ylabel('Number of Crime reported', fontsize=15)
plt.xlabel("'States/UTs')", fontsize=15)
plt.xticks(rotation=90)
plt.show()
plt.savefig('States1.jpg', transparent=True)
```

```python
data['States/UTs'].value_counts().plot.pie(autopct='%.2f',figsize=(30,30),fontsize=12)
plt.title("States's contributions towards total crime reports",fontsize=16)
plt.savefig('States_pie.jpg', transparent=True)
```
